```python
import plotly.graph_objs as go
fig = go.Figure(data=go.Heatmap(
    z=heart))

fig.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(heart,x='age',
                    y='trtbps', #resting blood pressure (in mm Hg)
                    z='thalachh', #maximum heart rate achieved
                    color='sex')
fig.show()
```

```python
from sklearn import linear_model
```

```python
x = heart[['trtbps','thalachh']]
y = heart['age']

regr = linear_model.LinearRegression()
regr.fit(x,y)

```

```python
predictHEART = regr.predict([[187,120]])
print(predictHEART)
```
