```python
def missing_cols(df):
  for col in df.columns:
    missing_vals = df[col].isnull().sum()
    if missing_vals !=0: 
      print(col)
      print(missing_vals)

missing_cols(messi)
```

```python
def prec_missing(df):
  for col in df.columns:
    pct = df[col].isnull().mean ()*100. #mean
    if pct !=0:
      print(col)
      print(pct)

prec_missing(messi)  
```

```python
nullMap = messi.isnull().replace(to_replace=[False,True],value=[0,1])
print(nullMap.head())
```

```python
fig = go.Figure(data=go.Heatmap(
    z=nullMap
))

fig.show()
```

```python
me1 = messi.drop(['Goal Type','Competition Name',],axis=1)
print(me1)
```

```python
display(me1.columns)
```

```python
me3 = me1.pivot_table(index = ['Final Game Result'],
                      columns =['Goal Method'],
                      values=[u'Score Home Team - When Messi Scored',u'Score Away Team - When Messi Scored'],
                      aggfunc='sum')
display(me3)

me4 = me1.pivot_table(index = ['Final Game Result'],
                      columns =['Goal Method'],
                      values=[u'Score Home Team - When Messi Scored',u'Score Away Team - When Messi Scored'],
                      aggfunc='mean')
display(me4)
```
