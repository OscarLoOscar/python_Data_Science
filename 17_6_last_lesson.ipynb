```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
immigration =  {
    'Age':[36,20,40,22,30],
    'Residence':[7,10,2,3,14],
    'Nationality':['UK','US','UK','SK','JP'],
    'Go':['No','Yes','No','No','Yes']
}

df = pd.DataFrame(immigration)
print(df)
```

```python
#用replace功能改data
n = {'UK':0,'US':1,'SK':2,'JP':3}
df['Nationality'] = df['Nationality'].replace(n)
print(df)
```

```python
#用replace功能改data
n = {'UK':0,'US':1,'SK':2,'JP':3}
df['Nationality'] = df['Nationality'].replace(n)
g = {"Yes":1,'No':0}
df["Go"]=df['Go'].replace(g)
print(df)
```

```python
features = ['Age' , 'Residence','Nationality']

x = df[features]
y = df['Go']

print(x)
print(y)
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)
data = tree.export_graphviz(dtree,feature_names = features)
```

```python
print(dtree.predict([[32,2,0]]))
```
