```python
import plotly.express as px

HAC = px.data.gapminder()
fig = px.choropleth(HAC, locations="Country",
                    color="HappinessScore", 
                    hover_name="Region", 
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

```python
import seaborn as sns
plt.figure(figsize = (15,6))
sns.barplot(x ='HappinessScore',y= 'Country',data = HAC.nlargest(25,'HappinessScore'), palette = 'viridis')
plt.title('Happiness Score of top 25 Countries', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
