```python
plt.figure(figsize = (15,6))
sns.barplot(x ='HappinessScore',y= 'Region',data = HAC, palette = 'gist_heat')
plt.title('Happiness Score by Region', fontsize = 20)
plt.xticks(rotation = 80)
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.barplot(x ='GDP_PerCapita',y= 'Region',data = HAC, palette = 'gist_heat')
plt.title('GDP by Region', fontsize = 20)
plt.xticks(rotation = 80)
plt.show()
```
