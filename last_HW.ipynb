```python
#Classification
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
data2 = data1.copy()
display(data2.head(2))
```

```python
labelencoder_Y = LabelEncoder()
data2.diagnosis = labelencoder_Y.fit_transform(data2.diagnosis)
```

```python
display(data2.head(2))
```

```python
data2.columns
```

```python
x = data2[['radius_mean','texture_mean','perimeter_mean','area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
y = data2['diagnosis']

regr = linear_model.LinearRegression()
regr.fit(x,y)

```

```python
print(data2.diagnosis.value_counts())
print("\n", data2.diagnosis.value_counts().sum())
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) #0.2=20% data 做 test
print(x_train)
```

```python
scaler = StandardScaler()
scaler.fit(x_train) 
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
print(x_train)
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
print(y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
error = []

for i in range(1,40):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(x_train,y_train)
  pred_i = knn.predict(x_test)
  error.append(np.mean(pred_i !=y_test))

print(error)
```

```python

plt.figure(figsize = (12,6))
plt.plot(range(1,40),error ,color='red',marker='o')
plt.xlabel('K value')
plt.ylabel('Mean Error')

#k value愈低愈準，所以大約7左右最準
```
