```python
import pandas as pd
```

```python
data1 = {
    'name': ['Amy','Benny','Cindy'],
    "score" :[70,40,90]
}

df1 = pd.DataFrame(data1)
print(df1)
```

```python
data2 = {
    "name": ['Amy','Benny','Danny'],
    'passed': [True,False,True]
}

df2 = pd.DataFrame(data2)
print(df2)
```

```python
#1st joining
df = df1.merge(df2,how = 'inner')
print(df)

#inner, outer, left, right
print()
df = df1.merge(df2,how = 'outer')
print(df) #NaN = Null ,data cleaning要處理
```

```python
df=df1.merge(df2,how ='left')
print(df)

df=df2.merge(df1,how ='left')
print(df)

df=df1.merge(df2,how ='right')
print(df)

df=df2.merge(df1,how ='right')
print(df)

#base on merge個句，邊個係前面就係left
```

```python
import plotly.express as px
```

```python
df = px.data.tips()
print(df)

df1 = df.pivot_table(index = 'sex')
print(df1)

df1 = df.pivot_table(index = 'sex',columns ='time')
print(df1)

df1 = df.pivot_table(index = 'sex',columns ='time',values='tip')
print(df1)

df1 = df.pivot_table(index = 'sex',columns ='time',values='total_bill')
print(df1)

df1 = df.pivot_table(index = 'sex',columns ='time',values='tip',aggfunc='mean')#defort左
print(df1)

#aggfunc影響點得出個值
df1 = df.pivot_table(index = 'sex',columns ='time',values='tip',aggfunc='sum')
print(df1)

df1 = df.pivot_table(index = 'sex',columns =['time','smoker'],values='tip',aggfunc='sum')
print(df1)

df1 = df.pivot_table(index = 'sex',columns =['time','smoker','day'],values='tip',aggfunc='sum')
print(df1)

df1 = df.pivot_table(index = ['sex','smoker'],columns =['time','day'],values='tip',aggfunc='sum')
print(df1)
```

```python
#背以下format
import plotly.graph_objs as go
airbnb_url = 'https://raw.githubusercontent.com/ManarOmar/New-York-Airbnb-2019/master/AB_NYC_2019.csv'
airbnb_ori = pd.read_csv(airbnb_url)
airbnb = airbnb_ori.copy()
airbnb.head()
```

```python
airbnb.info()
```

```python
def missing_cols(df):
  for col in df.columns:
    missing_vals = df[col].isnull().sum(). #isnull作用：將columns內的null轉成TRUE
    if missing_vals !=0: #true=0,flase=1
      print(col)
      print(missing_vals)

missing_cols(airbnb)
```

```python
def prec_missing(df):
  for col in df.columns:
    pct = df[col].isnull().mean ()*100. #mean
    if pct !=0:
      print(col)
      print(pct)

prec_missing(airbnb)  
```

```python
nullMap = airbnb.isnull().replace(to_replace=[False,True],value=[0,1])
print(nullMap.head())
```

```python
fig = go.Figure(data=go.Heatmap(
    z=nullMap
))

fig.show()
```
