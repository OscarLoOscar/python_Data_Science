```python
np.random.seed(2)

x = np.random.normal(3,1,100)
y = np.random.normal(150,40,100) / x

plt.scatter(x,y)
plt.show()
```

```python
train_x =x[:80] #cut開dataset，80做training，20做testing,避免overfit，underfit
train_y =y[:80]

test_x=x[80:]
test_y=y[80:]
#純粹係80:20，82比
```

```python
plt.scatter(train_x,train_y)
plt.scatter(test_x, test_y)
plt.show()
```

```python
mymodel = np.poly1d(np.polyfit(train_x,train_y,4)) #4係轉灣次數,ting / Underfitting

myline = np.linspace(0,6,500)

plt.scatter(train_x,train_y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
mymodel = np.poly1d(np.polyfit(train_x,train_y,3)) #3係轉灣次數,ting / Underfitting

myline = np.linspace(0,6,500)

plt.scatter(train_x,train_y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
mymodel = np.poly1d(np.polyfit(train_x,train_y,3)) #2係轉灣次數,ting / Underfitting

myline = np.linspace(0,6,500) #0,6指起點終點

plt.scatter(train_x,train_y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```
