```python
#重疊chat
from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y":True}]])

fig.add_trace(
    go.Scatter(x=[1,2,3],y=[40,50,60],name="yaxis"),
    secondary_y=False
)

fig.add_trace(
    go.Scatter(x=[2,3,4],y=[30,60,80],name="yaxis-2"),
    secondary_y=True
)

fig.show()
```

```python
#project management用

df = pd.DataFrame([
    dict(Task = "Job A",Start="2022-01-01",Finish="2022-05-01", Completion_pct=30),
    dict(Task = "Job B",Start="2021-01-01",Finish="2022-05-01", Completion_pct=50),
    dict(Task = "Job C",Start="2020-01-01",Finish="2022-05-01", Completion_pct=75),
    dict(Task = "Job D",Start="2019-01-01",Finish="2022-05-01", Completion_pct=100)
])

fig = px.timeline(df, x_start="Start", x_end="Finish",y="Task",color="Completion_pct")
fig.show()
```

```python
fig = go.Figure(go.Waterfall(                     #用waterfall代表有次序
    orientation = "v",                            #vetical,H=horesotal
    measure = ["","","total","","","total"],      #得兩樣野SET，一係TOTAL，一係唔TOTAL
    x = ["Affiliate","Direct Response","Net Inflow","Cancellation","Expired","Net Acquisition"],
    textposition = "outside",                     #宇放出放入分別
    text = [320,220,"",-50,-30,""],
    y = [320,220,"",-50,-30,""]
))

fig.update_layout(
    title = "Customer Acquisition 2021 Q1"
)

fig.show()

#打直由左邊睇起

```
