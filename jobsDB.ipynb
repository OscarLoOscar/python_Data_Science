```python
import bs4 as bs
import urllib.request
import json
```

```python


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

req = urllib.request.Request('https://hk.jobsdb.com/java-jobs?salaryrange=0-30000&salarytype=monthly&sortmode=ListedDate', headers=headers)

source = urllib.request.urlopen(req).read()

soup = bs.BeautifulSoup(source, 'lxml')

divs = soup.find_all('div', {'data-search-sol-meta': True})

articles = []

for div in divs:
    job_link = div.find('a', {'data-automation': 'job-list-view-job-link'})
    job_title = div.find('a', {'data-automation': 'jobTitle'})
    job_company = div.find('a', {'data-automation': 'jobCompany'})
    job_location = div.find('a', {'data-automation': 'jobLocation'})

    if job_link and job_title and job_company and job_location:
        # print(f"Job Link: https://hk.jobsdb.com{job_link.get('href')}")
        # print(f"Job Title: {job_title.text.strip()}")
        # print(f"Job Company: {job_company.text.strip()}")
        # print(f"Job Location: {job_location.text.strip()}")
        # print('-' * 80)
        articles.append({
            'job_link': f"https://hk.jobsdb.com{job_link.get('href')}",
            'job_title': job_title.text.strip(),
            'job_company': job_company.text.strip(),
            'job_location': job_location.text.strip()
        })

articles_json = json.dumps(articles, ensure_ascii=False, indent=4)
print(articles_json)
```
